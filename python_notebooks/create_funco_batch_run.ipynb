```python
import pandas as pd
import os
import csv
```

```python
vcf_path = '../new_vcf_100/'

files = os.listdir(vcf_path)
f_paths = [vcf_path + x for x in files]
cmds_list = []

for f in f_paths:
    cell_id = f.strip(vcf_path)
    cmd = 'gatk Funcotator -R ref/hg38-plus.fa -V ' + f + ' -O out/' + cell_id + '_benchmark.vcf \
    --data-sources-path funcotator_dataSources.v1.6.20190124g --ref-version hg38 -output-file-format VCF'
    #print(cmd)
    cmds_list.append(cmd)
```

```python
cmds_list
```

```python
new_cmds = []
for cmd in cmds_list:
    cmd1 = cmd.replace('../', '')
    print(cmd1)
    new_cmds.append(cmd1)
```

```python
new_cmds
```

```python
# write
with open("out.csv","w") as f:
    wr = csv.writer(f,delimiter="\n")
    wr.writerow(new_cmds)
```
